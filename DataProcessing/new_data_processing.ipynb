```python
import pandas as pd

df = pd.read_excel('new_data.xlsx', engine='openpyxl')
df = df.drop(columns=['Cost'], errors='ignore')
df.head()
```

```python
from faker import Faker
import pandas as pd

fake = Faker()

# Function to generate a birthdate for a given age
def get_birthdate(age):
    return fake.date_of_birth(minimum_age=age, maximum_age=age)

# Apply the function to each row in the Age column
df['Birthdate'] = df['Age'].apply(get_birthdate)

df.head()
```

```python
fake = Faker('nl_NL')  # Dutch locale

def get_first_name(gender):
    if gender == 'Male':
        return fake.first_name_male()
    if gender == 'Female':
        return fake.first_name_female()
    

def get_last_name():
    return fake.last_name()

# Assuming df is already loaded and has a 'Gender' column
df['First_Name'] = df['Gender'].apply(get_first_name)
df['Last_Name'] = df['Gender'].apply(lambda x: get_last_name())

df.head()
```

```python
# Reorder columns in df: patient_ID, First_Name, Last_Name, Gender, Birthdate, and the rest
cols = list(df.columns)
# Find the rest of the columns except the specified ones
main_cols = ['Patient_ID', 'First_Name', 'Last_Name', 'Gender', 'Birthdate']
rest_cols = [col for col in cols if col not in main_cols]
# Reorder
df = df[main_cols + rest_cols]

df.head()
```
