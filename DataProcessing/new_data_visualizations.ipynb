```python
#import pandas as pd
#import matplotlib.pyplot as plt
#import seaborn as sns

data = pd.read_csv('final_cleaned_data.csv')
data.head()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['Age'], bins=20, kde=True, color='skyblue')
plt.title('Age Distribution of Patients')
plt.xlabel('Age')
plt.ylabel('Number of Patients')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=data, x='Gender', hue='Gender', palette='pastel', legend=False)
plt.title('Gender Distribution of Patients')
plt.xlabel('Gender')
plt.ylabel('Number of Patients')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(14, 8))
sns.countplot(data=data, y='Condition', order=data['Condition'].value_counts().index,palette='viridis',legend=False)
plt.title('Frequency of Patient Conditions')
plt.xlabel('Number of Patients')
plt.ylabel('Condition')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(14, 8))
sns.countplot(data=data, y='Procedure', order=data['Procedure'].value_counts().index, palette='magma')
plt.title('Frequency of Medical Procedures')
plt.xlabel('Number of Patients')
plt.ylabel('Procedure')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='Length_of_Stay', y='Satisfaction', hue='Readmission', palette='coolwarm')
plt.title('Length of Stay vs. Satisfaction (Colored by Readmission)')
plt.xlabel('Length of Stay (Days)')
plt.ylabel('Satisfaction Score')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(14, 8))
readmission_data = data.groupby('Condition')['Readmission'].value_counts(normalize=True).unstack().fillna(0)
readmission_data.plot(kind='barh', stacked=True, figsize=(14, 8), color=['lightcoral', 'mediumseagreen'])
plt.title('Readmission Rates by Condition')
plt.xlabel('Proportion of Patients')
plt.ylabel('Condition')
plt.legend(title='Readmission', loc='lower right')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(16, 10))
sns.boxplot(data=data, y='Condition', x='Age', palette='Set3')
plt.title('Age Distribution per Condition')
plt.xlabel('Age')
plt.ylabel('Condition')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(16, 10))
sns.violinplot(data=data, y='Condition', x='Age', palette='Set2', inner='quartile', scale='width')
plt.title('Age Distribution per Condition (Violin Plot)')
plt.xlabel('Age')
plt.ylabel('Condition')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(14, 8))
contrast_data = data.groupby('Procedure')['Contrast'].value_counts(normalize=True).unstack().fillna(0)
contrast_data.plot(kind='barh', stacked=True, figsize=(14, 8), color=['steelblue', 'yellowgreen'])
plt.title('Contrast Usage per Procedure')
plt.xlabel('Proportion of Patients')
plt.ylabel('Procedure')
plt.legend(title='Contrast Used', labels=['No', 'Yes'], loc='lower right')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(16, 10))
procedure_condition = pd.crosstab(data['Condition'], data['Procedure'])

sns.heatmap(procedure_condition, annot=True, fmt='d', cmap='YlGnBu', linewidths=0.5)
plt.title('Correlation between Procedure and Condition (Frequency Heatmap)')
plt.xlabel('Procedure')
plt.ylabel('Condition')
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```
